```python
def extract_flight_route(review_content):
    # Regular expression patterns to match different formats
    pattern1 = r'\b([A-Z]{3}-[A-Z]{3})\b'
    pattern2 = r'\b([A-Z]{3} to [A-Z]{3})\b'


    # Try matching each pattern
    match1 = re.search(pattern1, review_content)
    match2 = re.search(pattern2, review_content)

    # Return the matched flight route
    if match1:
        return match1.group(1)
    elif match2:
        return match2.group(1).replace(' to ', '-')

    else:
        return None

# # Apply the function to create a new 'Flight_route' column
airline_review_3['Flight_route'] = ''
airline_review_3['Flight_route'] = airline_review['Review_content'].apply(extract_flight_route)

# # Replace 'None' values with an empty string ('')
# airline_review_3['Flight_Route'] = airline_review['Flight_Route'].fillna('')
# # Display the DataFrame with the new column
# # print(airline_review_3[['Review_title', 'Flight_route']])
```

A correlation matrix is a table showing correlation coefficients between variables. Each cell in the table represents the correlation between two variables. The value is in the range of -1 to 1, where:

1 indicates a perfect positive correlation,
-1 indicates a perfect negative correlation, and
0 indicates no correlation.


Here are some key points on how to interpret a correlation matrix:



Strength of Correlation:
The magnitude of the correlation coefficient indicates the strength of the relationship.
A coefficient close to 1 or -1 indicates a strong positive or negative correlation, respectively.
A coefficient close to 0 indicates a weak or no linear correlation.


Direction of Correlation:
A positive correlation (closer to 1) implies that as one variable increases, the other tends to increase.
A negative correlation (closer to -1) implies that as one variable increases, the other tends to decrease.


Correlation vs. Causation:
Correlation does not imply causation. Even if two variables are correlated, it does not necessarily mean that one causes the other.
There may be other factors influencing both variables or a third variable influencing them.


Interpreting Specific Values:
1: Perfect positive correlation (as one variable increases, the other increases proportionally).
-1: Perfect negative correlation (as one variable increases, the other decreases proportionally).
0: No correlation.
Values between 0 and 1 or 0 and -1 indicate the strength of correlation, with 0 being no correlation.


Useful for Feature Selection:
In machine learning, a correlation matrix is often used for feature selection. Highly correlated features may provide redundant information, and one of them may be removed to simplify the model.


Visualizing Correlations:
Heatmaps are commonly used to visualize correlation matrices. Darker colors (closer to 1 or -1) indicate stronger correlations.

#Hyothesis Testing

##Encoding

```python
encoded = pd.get_dummies(airline_review_clean[['Rating','Class']])
```

```python
encoded
```

```python
import scipy.stats as stats

# Shapiro-Wilk test for normality
alpha = 0.05
stat, p_value = stats.shapiro(encoded['Rating'])
print(f'Shapiro-Wilk Test - Statistic: {stat}')
print(f'p-value: {p_value}')

if p_value > alpha:
    print("The data appears to be normally distributed.")
else:
    print("The data may not be normally distributed.")


# Visualize the distribution using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(encoded['Rating'], kde=True, color='Orange')
plt.title('Histogram of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
import scipy.stats as stats
from scipy.stats import kruskal

# Perform Kruskal-Wallis test
h_statistic, p_value = kruskal(
    encoded['Rating'][encoded['Class_Business Class'] == 1],
    encoded['Rating'][encoded['Class_Economy Class'] == 1],
    encoded['Rating'][encoded['Class_First Class'] == 1],
    encoded['Rating'][encoded['Class_Premium Economy'] == 1]
)


print(f'Kruskal-Wallis Statistic: {h_statistic}')
print(f'P-Value: {p_value}')


alpha = 0.05


if p_value < alpha:
    print("REJECT the null hypothesis. At least one travel class has a different average rating compared to the others.")
else:
    print("ACCEPT the null hypothesis. There is no significant difference in ratings among different travel classes.")
```
